*Question 3: The Shifting Problem*

Imagine a straight row of 10 chairs with 10 students sitting next to each other without any empty seats.

**Deleting from the end (O(1))**: If the very last student gets up and leaves, no one else has to move. It is instant.



**Deleting from the middle (O(n))**: If the student in the middle gets up and leaves, it creates an empty gap. To fix this, every single student sitting to the right of that gap must slide over one seat to fill the empty space. In an array, elements must be stored in consecutive memory positions. Deleting from the middle forces the computer to move (shift) all the remaining elements backward, which takes a lot of time.

*Question 5: Array Traversal & Logic*

```python
def evenNumbers(array):
    for number in array:
        if number % 2 == 0:
            print(number)

myArray = [3, 8, 12, 5, 7, 14]
evenNumbers(myArray)
```

*Question 6: Linear Search Implementation*

```python
def linearSearch(array, target):
    for index in range(len(array)):
        if array[index] == target:
            return index
        
    return -1
        
test_array = [12, 4, 40, 20, 78]
print(linearSearch(test_array, 20))
print(linearSearch(test_array, 78))
print(linearSearch(test_array, 100))
```
