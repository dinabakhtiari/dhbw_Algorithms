*4.  LINEAR SEARCH*

**Algorithm:** `linearSearch(arr, target, n)`
> // This algorithm returns the index of a specific element in an array.

**Pseudocode:**
```javascript
Algorithm linearSearch(arr, target, n)
let x = 0; 
for (i = 0; i < n; i++) {
    if (arr[i] == target) {
        x = i;
    } // if it couldn't find the index, it returns -1
    else x = -1;
}
return x;

````python
# python implementation

arr = [12, 45, 67, 90, 3, 0, 4]
target = 67
def linearSearch(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


print("Index of target :", linearSearch(arr, target)) 
````

*6. VOWEL COUNTER*

**Algorithm:** `countVowels(str, n)`

> // This algorithm returns the total frequency of vowels found in a string

```javascript
Algorithm countVowels(str, n)

vowels = "aeiou"
count = 0
for(i = 1; i < n, i++){
    for(j = 1; j < 5; j++){
        if str[i] == vowels[j]
            count++;
    }
}
return count;

````python
# Python Implementation of Vowel Counter

text = "Dina and Mohamed"
def countVowels(text):
    vowels = "aeiou"
    count = 0

    text = text.lower()

    for char in text: 
        if char in vowels:
            count += 1
    return count

print("Total vowels: ", countVowels(text))
````
