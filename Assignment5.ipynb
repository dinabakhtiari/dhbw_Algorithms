Problem 1: The Stack (Reverse a String)

We use a Stack because it follows the LIFO rule

```python
def reverse_string(text):
    stack = []
    
    for char in text:
        stack.append(char)
        
    reversed_text = ""
    
    while len(stack) > 0:
        reversed_text += stack.pop()
        
    return reversed_text

print(reverse_string("hello"))  
```

Problem 2: The Queue (First Come, First Served)

We use a Queue because a ticketing office operates on a 'First-Come, First-Served' basis, which matches the FIFO rule, meaning 'First-In, First-Out'.

```python
class TicketQueue:
    def __init__(self):
        self.queue = []

    def add_customer(self, name):
        self.queue.append(name)

    def serve_customer(self):
        if len(self.queue) == 0:
            return "No customers in line"
        
        return self.queue.pop(0)
    

tq = TicketQueue()
tq.add_customer("Dina")
tq.add_customer("James")
tq.add_customer("Taha")
print(tq.serve_customer())
```

Problem 3: The Linked List (Find the Length)



```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def count_nodes(head):
    count = 0
    current = head

    while current is not None:
        count += 1
        current = current.next 
        
    return count

node1 = Node(10)
node2 = Node(20)
node1.next = node2
print(count_nodes(node1))
```
