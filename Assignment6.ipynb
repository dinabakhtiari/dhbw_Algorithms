```python
def isValidBST(root):

    def check(node, min, max):

        if not node:
            return True

        if not(min < node['val'] < max):
            return False

        return check(node['left'], min, node['val']) and check(node['right'], node['val'], max)

    return check(root, float('-inf'), float('inf'))


tree_1 = {
    'val': 10,
    'left': {'val': 5, 'left': None, 'right': None},
    'right': {'val': 15, 'left': None, 'right': None}
}

tree_2 = {
    'val': 10,
    'left': {'val': 5, 'left': None, 'right': None},
    'right': {'val': 3, 'left': None, 'right': None}
}

print("Is tree_1 a BST?", isValidBST(tree_1))
print("Is tree_2 a BST?", isValidBST(tree_2))
```
